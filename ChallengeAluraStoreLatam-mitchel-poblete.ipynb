### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



# 1.1 Ingreso total por cada tienda

```python
# Combina las bases de datos en un solo DataFrame para analisis mas completo --- ignore_index=True PARA QUE CADA VALOR DE LA TABLA TENGA UN INDICE UNICO Y NO SE REPITA
tiendas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)


# Definicion de la funcion para calcular el ingreso total por 'Lugar de Compra'
def calcular_ingreso_total(df):
    # Agrupar los datos por 'Lugar de Compra' y calcular la suma del 'Precio'
    ingreso_total = df.groupby('Lugar de Compra')['Precio'].sum().reset_index()

    # Renombrar las columnas del DataFrame resultante
    ingreso_total.columns = ['Lugar de Compra', 'Ingreso Total']

    # Ordenar el DataFrame por 'Ingreso Total' de mayor a menor
    ingreso_total = ingreso_total.sort_values(by='Ingreso Total', ascending=False)

    # Devolver el DataFrame resultante
    return ingreso_total

# Definición de la función para calcular el ingreso total de todas las tiendas
def total_lugares_de_compras(df):

    # Calcular la suma total de la columna 'Precio'
    ingreso_total = df['Precio'].sum()

    # Devolver el ingreso total
    return ingreso_total


# Calcular el ingreso total por 'Lugar de Compra' para todas las tiendas combinadas
ganancias_tienda1_total = total_lugares_de_compras(tienda)
ganancias_tienda2_total = total_lugares_de_compras(tienda2)
ganancias_tienda3_total = total_lugares_de_compras(tienda3)
ganancias_tienda4_total = total_lugares_de_compras(tienda4)

# Calcular el ingreso total por 'Lugar de Compra' para cada tienda
ingreso_total_tienda = calcular_ingreso_total(tienda)
ingreso_total_tienda2 = calcular_ingreso_total(tienda2)
ingreso_total_tienda3 = calcular_ingreso_total(tienda3)
ingreso_total_tienda4 = calcular_ingreso_total(tienda4)


# Crea graficos de barras para cada tienda
def plot_ingreso_total(ingreso_total, title, total_ingreso):#ingreso_total; contiene el ingreso total por 'Lugar de Compra'.   title:titulo del grafico  total_ingreso:ingreso total de la tienda.
    plt.figure(figsize=(10, 6)) #Crea una nueva figura con un tamaño de 10 pulgadas de ancho y 6 pulgadas de alto
    plt.bar(ingreso_total['Lugar de Compra'], ingreso_total['Ingreso Total'], color='skyblue')#plt.bar: crea un grafico de barras. X e Y
    plt.xlabel('Lugar de Compra')# plt.xlabel('Lugar de Compra') etiqueta X
    plt.ylabel('Ingreso Total')#plt.ylabel('Ingreso Total') etiqueta Y
    plt.title(title)#Titulo del grafico que se pasas como parametro
    plt.xticks(rotation=45)#rota los valores de X para que se lea de mejor manera
    plt.tight_layout() #ajusta los elementos para que no se superpongan
    plt.text(0.5, 0.95, f'Ganancias Totales: ${int(total_ingreso):,}', transform=plt.gca().transAxes, ha='center')#agrega texto al grafico (0.5, 0.95 Posicion del texto en coordenadas(centro superior del grafico)
                                                                      # transform=plt.gca().transAxes, ha='center' toma los ejes ya ingresados que son(0.5, 0.95 y centra el texto "ganancias totales")
    plt.show()


# Mostrar los resultados
print("Ingreso total por 'Lugar de Compra' para tienda1:")
print(ingreso_total_tienda)
print(" \n----GANANCIAS TOTALES : $",int(ganancias_tienda1_total),"----")
# Llamar a la función para cada tienda
plot_ingreso_total(ingreso_total_tienda, "Ingreso Total por Lugar de Compra - Tienda 1", ganancias_tienda1_total)
print("\n")

print("\nIngreso total por 'Lugar de Compra' para tienda2:")
print(ingreso_total_tienda2)
print(" \n----GANANCIAS TOTALES : $",int(ganancias_tienda2_total),"----")
plot_ingreso_total(ingreso_total_tienda2, "Ingreso Total por Lugar de Compra - Tienda 2", ganancias_tienda2_total)
print("\n")

print("\nIngreso total por 'Lugar de Compra' para tienda3:")
print(ingreso_total_tienda3)
print(" \n----GANANCIAS TOTALES : $",int(ganancias_tienda3_total),"----")
plot_ingreso_total(ingreso_total_tienda3, "Ingreso Total por Lugar de Compra - Tienda 3", ganancias_tienda3_total)
print("\n")

print("\nIngreso total por 'Lugar de Compra' para tienda4:")
print(ingreso_total_tienda4)
print(" \n----GANANCIAS TOTALES : $",int(ganancias_tienda4_total),"----")
plot_ingreso_total(ingreso_total_tienda4, "Ingreso Total por Lugar de Compra - Tienda 4", ganancias_tienda4_total)
print("\n")

```

# 2. Ventas por categoría

```python
# Definición de la funcion
def calcular_ventas_por_categoria(df):
    # Agrupar los datos por 'Categoría del Producto' y calcular la suma del 'Precio'
    ventas_por_categoria = df.groupby('Categoría del Producto')['Precio'].sum().reset_index()

    # Renombrar las columnas del DataFrame resultante
    ventas_por_categoria.columns = ['Categoria', 'Ventas Totales']

    # Ordenar el DataFrame por 'Ventas Totales' de mayor a menor
    ventas_por_categoria = ventas_por_categoria.sort_values(by='Ventas Totales', ascending=False)

    # Devolver el DataFrame resultante
    return ventas_por_categoria



# Crear graficos de pastel para cada tienda
def plot_ventas_por_categoria(ventas_por_categoria, title):
    plt.figure(figsize=(10, 6))
    plt.pie(ventas_por_categoria['Ventas Totales'], labels=ventas_por_categoria['Categoria'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
    plt.title(title)
    plt.tight_layout()
    plt.xticks(rotation=45)#rota los valores de X para que se lea de mejor manera
    plt.show()


# Calcular las ventas por categoría para cada tienda
ventas_por_categoria_tienda = calcular_ventas_por_categoria(tienda)
ventas_por_categoria_tienda2 = calcular_ventas_por_categoria(tienda2)
ventas_por_categoria_tienda3 = calcular_ventas_por_categoria(tienda3)
ventas_por_categoria_tienda4 = calcular_ventas_por_categoria(tienda4)

# Mostrar los resultados
print("Ventas por categoría para tienda1:")
print(ventas_por_categoria_tienda)
# Llamar a la función para cada tienda
plot_ventas_por_categoria(ventas_por_categoria_tienda, "Ventas por Categoría - Tienda 1")
print("\n")
print("\nVentas por categoría para tienda2:")
print(ventas_por_categoria_tienda2)
plot_ventas_por_categoria(ventas_por_categoria_tienda2, "Ventas por Categoría - Tienda 2")
print("\n")
print("\nVentas por categoría para tienda3:")
print(ventas_por_categoria_tienda3)
plot_ventas_por_categoria(ventas_por_categoria_tienda3, "Ventas por Categoría - Tienda 3")
print("\n")
print("\nVentas por categoría para tienda4:")
print(ventas_por_categoria_tienda4)
plot_ventas_por_categoria(ventas_por_categoria_tienda4, "Ventas por Categoría - Tienda 4")





```

# 3. Calificación promedio de la tienda


```python
# Definicion de la funcion para calcular la calificacion promedio por tienda
def calcular_calificacion_promedio(df):
    # Agrupar los datos por 'Lugar de Compra' y calcular el promedio de la 'Calificacion'
    calificacion_promedio = df.groupby('Lugar de Compra')['Calificación'].mean().reset_index()

    # Redondear el promedio a 2 decimales
    calificacion_promedio['Calificación'] = calificacion_promedio['Calificación'].round(2)

    # Renombrar las columnas del DataFrame resultante
    calificacion_promedio.columns = ['Tienda', 'P.C']

    # Ordenar el DataFrame por 'Calificación Promedio' de mayor a menor
    # sort_values ordena los valores en orden ascendente (ascending=True), pero si se establece ascending=False, los valores se ordenaran en orden descendente.
    calificacion_promedio = calificacion_promedio.sort_values(by='P.C', ascending=False)

    # Devolver el DataFrame resultante
    return calificacion_promedio


# Crear graficos de puntos para cada tienda
def plot_calificacion_promedio(calificacion_promedio, title):
    plt.figure(figsize=(10, 6))
    plt.scatter(calificacion_promedio['Tienda'], calificacion_promedio['P.C'], color='skyblue', s=100, edgecolor='black')
    plt.xlabel('Tienda')
    plt.ylabel('Calificación Promedio')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True)
    plt.show()


# Calcular la calificación promedio para cada tienda
calificacion_promedio_tienda = calcular_calificacion_promedio(tienda)
calificacion_promedio_tienda2 = calcular_calificacion_promedio(tienda2)
calificacion_promedio_tienda3 = calcular_calificacion_promedio(tienda3)
calificacion_promedio_tienda4 = calcular_calificacion_promedio(tienda4)

# Mostrar los resultados
print("Calificacion promedio para tienda:")
print(calificacion_promedio_tienda)
plot_calificacion_promedio(calificacion_promedio_tienda, "Calificacion Promedio por Tienda - Tienda 1")
print("\n")
print("\nCalificacion promedio para tienda2:")
print(calificacion_promedio_tienda2)
plot_calificacion_promedio(calificacion_promedio_tienda2, "Calificacion Promedio por Tienda - Tienda 2")
print("\n")
print("\nCalificacion promedio para tienda3:")
print(calificacion_promedio_tienda3)
plot_calificacion_promedio(calificacion_promedio_tienda3, "Calificacion Promedio por Tienda - Tienda 3")
print("\n")
print("\nCalificacion promedio para tienda4:")
print(calificacion_promedio_tienda4)
plot_calificacion_promedio(calificacion_promedio_tienda4, "Calificacion Promedio por Tienda - Tienda 4")


```

# 4. Productos más y menos vendidos

```python
# Definicion de la funcion para calcular los productos mas y menos vendidos por tienda
def calcular_productos_vendidos(df):
    # Agrupar los datos por 'Producto' y calcular la suma de la 'Cantidad de cuotas' lo hice asi por que (NO HAY CANTIDAD DE VENTAS EN LA BD)
    productos_vendidos = df.groupby('Producto')['Cantidad de cuotas'].sum().reset_index()

    # Encontrar el producto mas vendido
    producto_mas_vendido = productos_vendidos.loc[productos_vendidos['Cantidad de cuotas'].idxmax()]#.idxmax() Devuelve el primer valor maximo en una columna

    # Encontrar el producto menos vendido
    producto_menos_vendido = productos_vendidos.loc[productos_vendidos['Cantidad de cuotas'].idxmin()]#.idxmin() Devuelve el ultimo valor minimo en una columna

    # Crear un DataFrame con los resultados . pd.DataFrame({'': ['','']}) esta funcion de pandas se usa para crear un DataFrame.
    resultados = pd.DataFrame({
        'Producto': ['Mas Vendido', 'Menos Vendido'],
        'Nombre': [producto_mas_vendido['Producto'], producto_menos_vendido['Producto']],
        'Ventas': [producto_mas_vendido['Cantidad de cuotas'], producto_menos_vendido['Cantidad de cuotas']]
    })

    return resultados

# Crear graficos de líneas con marcadores para cada tienda
def plot_productos_vendidos(productos_vendidos, title):
    plt.figure(figsize=(10, 6))
    plt.plot(productos_vendidos['Producto'], productos_vendidos['Ventas'], marker='o', linestyle='-', color='skyblue', markeredgecolor='black', markersize=10)
    plt.xlabel('Producto')
    plt.ylabel('Ventas')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True)
    plt.show()
    plt.show()

# Calcular los productos mas y menos vendidos para cada tienda
productos_vendidos_tienda = calcular_productos_vendidos(tienda)
productos_vendidos_tienda2 = calcular_productos_vendidos(tienda2)
productos_vendidos_tienda3 = calcular_productos_vendidos(tienda3)
productos_vendidos_tienda4 = calcular_productos_vendidos(tienda4)

# Mostrar los resultados
print("Productos mas y menos vendidos para tienda1:")
print(productos_vendidos_tienda)
plot_productos_vendidos(productos_vendidos_tienda, "Productos más y menos vendidos - Tienda 1")
print("\n")
print("\nProductos mas y menos vendidos para tienda2:")
print(productos_vendidos_tienda2)
plot_productos_vendidos(productos_vendidos_tienda2, "Productos más y menos vendidos - Tienda 2")
print("\n")
print("\nProductos mas y menos vendidos para tienda3:")
print(productos_vendidos_tienda3)
plot_productos_vendidos(productos_vendidos_tienda3, "Productos más y menos vendidos - Tienda 3")
print("\n")
print("\nProductos mas y menos vendidos para tienda4:")
print(productos_vendidos_tienda4)
plot_productos_vendidos(productos_vendidos_tienda4, "Productos más y menos vendidos - Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Definicion de la funcion para calcular el costo de envio promedio por tienda
def calcular_costo_envio_promedio(df):
    # Agrupar los datos por 'Lugar de Compra' y calcular el promedio del 'Costo de envio'
    costo_envio_promedio = df.groupby('Lugar de Compra')['Costo de envío'].mean().reset_index()

    # Redondear el promedio a 2 decimales
    costo_envio_promedio['Costo de envío'] = costo_envio_promedio['Costo de envío'].round(2)

    # Renombrar las columnas del DataFrame resultante
    costo_envio_promedio.columns = ['Lugar de Compra', 'C.P.E']

    # Ordenar el DataFrame por 'Costo de Envío Promedio' de mayor a menor (by='C.P.E') Especifica la columna por la cual se quiere ordenar el DataFrame
    costo_envio_promedio = costo_envio_promedio.sort_values(by='C.P.E', ascending=False)

    # Devolver el DataFrame resultante
    return costo_envio_promedio

# Crear graficos de barras horizontales para cada tienda
def plot_costo_envio_promedio(costo_envio_promedio, title):
    plt.figure(figsize=(10, 6))
    plt.barh(costo_envio_promedio['Lugar de Compra'], costo_envio_promedio['C.P.E'], color='skyblue')
    plt.xlabel('Costo de Envío Promedio')
    plt.ylabel('Lugar de Compra')
    plt.title(title)
    plt.tight_layout()
    plt.show()


# Calcular el costo de envio promedio para cada tienda
costo_envio_promedio_tienda = calcular_costo_envio_promedio(tienda)
costo_envio_promedio_tienda2 = calcular_costo_envio_promedio(tienda2)
costo_envio_promedio_tienda3 = calcular_costo_envio_promedio(tienda3)
costo_envio_promedio_tienda4 = calcular_costo_envio_promedio(tienda4)

# Mostrar los resultados
print("Costo de envio promedio para tienda1:")
print(costo_envio_promedio_tienda)
plot_costo_envio_promedio(costo_envio_promedio_tienda, "Costo de Envío Promedio por Lugar de Compra - Tienda 1")
print("\n")
print("\nCosto de envio promedio para tienda2:")
print(costo_envio_promedio_tienda2)
plot_costo_envio_promedio(costo_envio_promedio_tienda2, "Costo de Envío Promedio por Lugar de Compra - Tienda 2")
print("\n")
print("\nCosto de envio promedio para tienda3:")
print(costo_envio_promedio_tienda3)
plot_costo_envio_promedio(costo_envio_promedio_tienda3, "Costo de Envío Promedio por Lugar de Compra - Tienda 3")
print("\n")
print("\nCosto de envio promedio para tienda4:")
print(costo_envio_promedio_tienda4)
plot_costo_envio_promedio(costo_envio_promedio_tienda4, "Costo de Envío Promedio por Lugar de Compra - Tienda 4")
```
